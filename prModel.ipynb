```
import pyterrier as pt
if not pt.started():
  pt.init()
```

```
data = []
with open('/content/drive/MyDrive/2019qrels-pass.txt', 'rt') as txtfile:
  line=txtfile.readlines()
  for i in line:
    k = i.strip().split(' ')
    if int(k[3]) != 0:
      data.append(i)
print(data)
file_write_obj = open("2019qrels-pass-eval.txt", 'w')   
for var in data:
    file_write_obj.write(var) 
file_write_obj.close()
```

```
index = pt.IndexFactory.of("/content/drive/MyDrive/index/data.properties")
print(index.getCollectionStatistics().toString())
```

```
topics_path = "/content/drive/MyDrive/msmarco-test2019-queries.txt"
topics = pt.io.read_topics(topics_path, format='singleline')
topics.head(5)
```

```

qrels = pt.io.read_qrels('/content/drive/MyDrive/2019qrels-pass.txt')
```

```

BM25_br = pt.BatchRetrieve(index, wmodel="BM25")
res = BM25_br.transform(topics)
pt.Utils.evaluate(res, qrels, metrics = ['map','ndcg_cut_10','recip_rank'])
```

```
BM25_br.saveResult(res,"result1.res")
```

```
eval = pt.Utils.evaluate(res,qrels,metrics=['recip_rank'])
eval
```

```
eval = pt.Utils.evaluate(res,qrels,metrics=["map"], perquery=True)
eval
```

```
res
```

```
bm25_fb = pt.BatchRetrieve(index, wmodel="BM25", controls={"qe" : "on"})
res_fb = bm25_fb.transform(topics)
pt.Utils.evaluate(res_fb, qrels, metrics = ['map','ndcg_cut_10','recip_rank'])
```

```
pt.Experiment(
    [bm25, bm25_fb],
    topics,
    qrels,
    eval_metrics=["map", "recip_rank","ndcg_cut_20"],
    names=["bm25", "bm25_fb"],
    baseline=0
)
```

```
_res_fb
bm25_fb.saveResult(res_fb,"result_fb.res")
```

```
eval_fb = pt.Utils.evaluate(res_fb,qrels,metrics=["map"], perquery=True)
eval_fb
```

```
bm25 = pt.BatchRetrieve(index, wmodel="BM25")
dph = pt.BatchRetrieve(index, wmodel="DPH")
linear = bm25 + dph
```

```
res_li = linear.transform(topics)
pt.Utils.evaluate(res_li, qrels, metrics = ['map','ndcg_cut_10','recip_rank'])
```

```
eval_li = pt.Utils.evaluate(res_li,qrels,metrics=["map"], perquery=True)
eval_li
```
